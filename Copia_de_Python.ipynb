## 1. VARIABLES I TIPUS

Quan treballem en models de machine learning, en general, trobem aquests tipus de dades:
- **int/float** --> valors numèrics (int per enters i float per nombres reals)
- **bool** --> True/False, només dos possibles valors. Pot ser que ens els trobem codificats, amb True representat per 1 i False per 0. En aquest cas, el més senzill és tractar la variable com enters.
- **str** --> Dades de text
- **None** --> Valors absents

````python
# --- EXEMPLE 1.1: Tipus bàsics ---

# Dades d'un pacient
edat = 45
pes = 78.5
te_diabetis = True
grup_sanguini = "A+"
colesterol = None

print(f"Edat: {edat} → tipus: {type(edat).__name__}")
print(f"Pes: {pes} → tipus: {type(pes).__name__}")
print(f"Té diabetis: {te_diabetis} → tipus: {type(te_diabetis).__name__}")
print(f"Grup sanguini: {grup_sanguini} → tipus: {type(grup_sanguini).__name__}")
print(f"Colesterol: {colesterol} → tipus: {type(colesterol).__name__}")
````

````python
# --- EXEMPLE 1.2: Convertir tipus de dades ---

te_diabetis = True
te_diabetis_int = int(te_diabetis)
te_diabetis_float = float(te_diabetis)
te_diabetis_str = str(te_diabetis)
print(f"Té diabetis (original): {te_diabetis} → tipus: {type(te_diabetis).__name__}")
print(f"Té diabetis (convertit a enter): {te_diabetis_int} → tipus: {type(te_diabetis_int).__name__}")
print(f"Té diabetis (convertit a real): {te_diabetis_float} → tipus: {type(te_diabetis_float).__name__}")
print(f"Té diabetis (convertit a string): {te_diabetis_str} → tipus: {type(te_diabetis_str).__name__}")
````

Altres tipus de dades que podem trobar-nos, tot i que no són habituals, son les tuples i els conjunts:
- **Tupla** --> Estructura ordenada i immutable. Dins una tupla hi podem trobar diferents tipus de dades.
- **Conjunt** --> Estructura mutable, sense ordre i d'elements únics.


````python
# --- EXEMPLE 1.3: Tuples i conjunts ---

# TUPLA (tuple)
tuple = (1, "Hola", 3.14, True)
print(f"Tupla: {tuple} → tipus: {type(tuple).__name__} \n")

# CONJUNT (set)
animals = set(["dog", "cat", "bird", "cat", "bird", "horse", "horse"])
print(f"Animals: {animals} → tipus: {type(animals).__name__} \n")
````

## 2. LLISTES I DICCIONARIS

- Llista --> Estructura mutable, sense ordre i amb possibles elements repetits.
- Diccionari --> Estructura que emmagatzema dades en format clau-valor. Les claus han de ser úniques.

````python
# --- EXEMPLE 2.1: Operacions bàsiques amb llistes ---

fruit = ['apple', 'pear', 'melon', 'melon', 'apple', 'apple', 'banana']
print(f"Fruita: {fruit} → tipus: {type(fruit).__name__} \n")
print(f"Primer element de la llista: {fruit[0]}")
print(f"Segon element de la llista: {fruit[1]}")
print(f"Últim element de la llista: {fruit[-1]} \n")

print(f"Màxim element de la llista: {max(fruit)} \n")
print(f"Mínim element de la llista: {min(fruit)} \n")

fruit.append('kiwi') # Afegim kiwi a la llista
print(f"Afegir un element a la llista: {fruit}")

fruit.remove('melon') # Eliminem melon de la llista (s'elimina la primera instància)
print(f"Eliminar un element de la llista (melon): {fruit}")
````

Quan apliquem la funció max o min en una llista de paraules (str) ens torna com a element màxim la última paraula en ordre alfabètic i com a element mínim la primera paraula en ordre alfabètic. Si volem trobar quina és la paraula més llarga o més curta hem d'especificar el criteri ```key = len ```.

````python
long_fruit = max(fruit, key=len)
print(f"Paraula més llarga de la llista: {long_fruit}")

short_fruit = min(fruit, key=len)
print(f"Paraula més curta de la llista: {short_fruit}")
# Tot i que 'pear' i 'kiwi' tenen la mateixa llargada, ens retorna 'pear' perquè es la que hi ha abans a la llista
````

````python
# --- EXEMPLE 2.2: Operacions bàsiques amb diccionaris ---

dades = {
  "edat" : 45,
  "pes" : 78.5,
  "te_diabetis" : True,
  "grup_sanguini" : "A+",
  "colesterol" : None
}

print(f"Dades: {dades}")
print(f"Grup sanguini: {dades['grup_sanguini']}") # Accedir a un valor concret
print(f"Claus: {dades.keys()}")
print(f"Valors: {dades.values()} \n")

dades['alçada'] = 1.81            # Afegir una nova entrada al diccionari
dades['grup_sanguini'] = "O+"     # Modificar un valor del diccionari
dades['sexe'] = "M"
print(f"Dades: {dades}")
````

Podem combinar dues llistes i tractar-les com un "diccionari":

````python
# --- EXEMPLE 2.3: Combinacions de llistes ---

edats = [10, 25, 3, 88, 47, 62, 15, 99]

etiquetes = [1 if v >= 18 else 0 for v in edats] # Establim 1 pels majors d'edat i 0 pels menors

parells = list(zip(edats, etiquetes))
print(f"Parells (edat, etiqueta): {parells}")
````

Realment, el que acabem de fer és iterar per cada valor de la llista 'edats'. És a dir, hem usat un bucle.

## 3. BUCLES I FUNCIONS

Si desmuntem el que hem fet a l'apartat anterior, podem veure clarament l'estructura del bucle:

````python
# --- EXEMPLE 3.1: Bucle per a combinar dues llistes ---

edats = [10, 25, 3, 88, 47, 62, 15, 99]
etiquetes = []

for edat in edats:
  if edat >= 18:
    etiquetes.append(1)
  else:
    etiquetes.append(0)

parells = list(zip(edats, etiquetes))
print(f"Parells (edat, etiqueta): {parells}")
````

Una altra manera d'iterar és fent-ho de manera limitada. És a dir, només iterant un determinat nombre de vegades.

````python
# --- EXEMPLE 3.2: Iterar x vegades ---

x = 5
for i in range(x):
  print(i)
````

````python
# --- EXEMPLE 3.3: Iterar definint un inici i un final ---

inici = 2
final = 8       # No inclòs

for i in range(inici, final):
    print(i)
````

````python
# --- EXEMPLE 3.4: Iterar amb salts ---

inici = 2
final = 8       # No inclòs
pas = 2

for i in range(inici, final, pas):
    print(i)
````

````python
# --- EXEMPLE 3.5: Bucle descendent ---
# L'inici ha de ser més gran que el final

inici = 8
final = 2       # No inclòs

for i in range(inici, final, -1):
    print(i)
````

````python
# --- EXEMPLE 3.6: Iterar pels valors d'una llista ---

colors = ["groc", "blau", "verd"]
for i in range(len(colors)):
    print(f"Índex {i}: {colors[i]}")
````

Una funció és un bloc de codi ordenat i reutilitzable que realitza una o diverses tasques concretes. Es defineixen amb la paraula ```def```, seguida del nom de la funció, uns parèntesis () amb les variables input i dos punts :. Opcionalment, abans dels punts es pot especificar quin tipus de variable retornarà la funció, així com també es poden explicitar el tipus de dades d'entrada.

````python
# --- EXEMPLE 3.7: Funcions bàsiques ---

def sumar(a:float, b:float) -> float:
  return a + b

def restar(a, b):
  return a - b

suma = sumar(7,3)
resta = restar(7,3)
print(f"Suma = {suma} \n")
print(f"Resta = {resta} \n")

suma2 = sumar(suma, resta)  # Podem usar la funció tantes vegades com vulguem
print(f"Suma de la suma i la resta = {suma2}")
````

````python
# --- EXEMPLE 3.8: Altres funcions i bucles ---

def restar_min(llista):
  """Resta a cada valor d'una llista el mínim de la llista"""

  if not llista:
        return []

  minim = min(llista)

  # Creem una nova llista restant el màxim a cada element 'x'
  return [x - minim for x in llista]

dades = [10, 20, 30, 40, 50]
dades2 = restar_min(dades)

print(f"Llista original: {dades}")
print(f"Llista després de passar-hi la funció: {dades2}")
````

````python
# --- EXEMPLE 3.9: Funcions que retornen diversos valors ---

def max_min_mitj(llista):
  mitjana = sum(llista)/len(llista)
  return max(llista), min(llista), mitjana


dades = [10, 20, 30, 40, 50]
maxim, minim, mitjana = max_min_mitj(dades)

print(f"Màxim: {maxim} \n")
print(f"Mínim: {minim} \n")
print(f"Mitjana: {mitjana} \n")
````

## 4. Pandas i MatplotLib

Pandas i Matplotlib són llibreries imprescindibles tant per la construcció de models com pel tractament i la visualització de dades. Per a poder usar-les, el primer que cal fer és importar-les.

```import pandas as pd```

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# --- EXEMPLE 4.1: Pandas — DataFrame bàsic ---

# Creació d'un DataFrame
dades_dict = {
    "edat":     [25, 34, 45, 22, 38, 52, 29],
    "salari":   [28000, 45000, 62000, 22000, 51000, 75000, 33000],
    "estudis":  ["FP", "Grau", "Màster", "FP", "Grau", "Màster", "FP"],
    "compra":   [0, 1, 1, 0, 1, 1, 0]   # target: va comprar el producte?
}

df = pd.DataFrame(dades_dict)
print("DataFrame:")
print(df)
print(f"\nForma: {df.shape}")
print("\nInformació:")
df.info()
````

En la gran majoria dels casos, ens trobarem que les dades estan emmagatzemades en un fitxer independent, en format csv o excel. Si es així, podem importar les dades fent servir una d'aquestes funcions ja determinades a Pandas:
- ```df = pd.read_csv('PATH DEL FITXER CSV')```
- ```df = pd.read_excel('PATH DEL FITXER EXCEL')```

````python
# --- EXEMPLE 4.2: Pandas — exploració i preprocessament ---

print("Estadística descriptiva:")
print(df.describe())

print("Primeres 3 files")
print(df.head(3))

# Selecció de columnes
print("\nColumna edat:")
print(df["edat"])

# Filtrar files
print("\nPersones amb salari > 40000:")
print(df[df["salari"] > 40000])

# Valors nuls (fonamental en ML!)
df_amb_nuls = df.copy()
df_amb_nuls.loc[2, "edat"] = None
df_amb_nuls.loc[5, "salari"] = None
print("\nValors nuls per columna:")
print(df_amb_nuls.isnull().sum())
````

La llibreria Matplotlib la farem servir, sobretot, per a generar gràfiques de les nostres dades. Els exemples següents s'han dibuixat amb datasets no reals, construïts amb la llibreria NumPy. Per això, també caldrà importar-la.

En aquest enllaç trobareu altres exemples de visualitzacions, així com els codis necessaris per generar-los: https://matplotlib.org/stable/plot_types/index.html


````python
# --- EXEMPLE 4.3: Matplotlib — Barplot ---

import numpy as np


# Dades
x = 0.5 + np.arange(8)                           # [0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5, 7.5]
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

# Plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),     # Coordenada horitzontal
       ylim=(0, 8), yticks=np.arange(1, 8))     # Coordenada vertical

plt.show()
````

````python
# --- EXEMPLE 4.4: Matplotlib — Scatterplot ---


# Dades aleatòries
np.random.seed(42)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))

# Mida i color
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# Plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
````

````python
# --- EXEMPLE 4.5: Matplotlib — Pie Plot ---



# Dades
x = [1, 2, 3, 4]
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

# Plot
fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
````
